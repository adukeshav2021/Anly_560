```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('/content/lab5.csv')
data.head()
```

```python
pattern = "(\d{2})\s([\s\S]+)"
matches = data['department_title'].str.extract(pattern)
data['department_no'] = matches[0].astype(int)
data['department_title'] = matches[1]
```

```python
matches = data['character_title'].str.extract(pattern)
data['character_no'] = matches[0].astype(int)
data['character_title'] = matches[1]
```

```python
pattern = "^([\s\S]+)\s(\d{4}|\d{4}[A-Z])$"
matches = data['sub_obj_title'].str.extract(pattern)
data['sub_obj_code'] = matches[1]
data['sub_obj_title'] = matches[0]
```

```python
data.head()
```

```python
pattern = "^(\w{4})"
matches = data['document_no'].str.extract(pattern)
data['document_type'] = matches[0]
data.head()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df1 = data.groupby(['department_title'],as_index=False).agg({'transaction_amount': ['sum']})
#df1
df1 = df1.sort_values(by=[('transaction_amount','sum')],ascending=False,ignore_index=True)

df1.head()
```

```python
who = df1['department_title'][0]
df2 = data[data['department_title']==who]
df2.head()
```

```python
df3 = df2.groupby(['document_type'],as_index=False).agg({'transaction_amount': ['sum']})
#df1
df3 = df3.sort_values(by=[('transaction_amount','sum')],ascending=False,ignore_index=True)

df3.head()
```

```python
df4 = data[data['department_title']=='CITY COUNCIL']
df4 = df4[df4['doc_ref_no_prefix_definition']=='petty cash']
df4.head()
```

```python
df5 = df4.groupby(['vendor_name'],as_index=False).agg({'transaction_amount': ['sum']})
#df1
df5 = df5.sort_values(by=[('transaction_amount','sum')],ascending=False,ignore_index=True)

df5.head()
```
