```python
%%R
chisq <- function(inputData){ 
  
  i = 1
  data_sum = 0
  while(i <= length(inputData)) {
    data_sum = data_sum + inputData[i]
    i = i + 1
  }

  print(data_sum)

  rows_sum<- numeric(nrow(inputData))
  for (i in seq(ncol(inputData))) {
     rows_sum[i] <- sum(inputData[i, ])
      }

  print(rows_sum)

  cols_sum <- numeric(ncol(inputData))
  for (i in seq(ncol(inputData))) {
      cols_sum[i] <- sum(inputData[, i])
      }

  print(cols_sum)

  expected_values <-  matrix(1:10,nrow=2, ncol=2, byrow=T)

  for (i in 1:nrow(inputData)){
     eachCell <- matrix(1:10,nrow=2, ncol=2, byrow=T)
     for (j in 1:ncol(inputData)){
         
         eachCell[[i,j]] <- (rows_sum[i] * cols_sum[j] / data_sum) 
     }
    }
    print(eachCell)
  expected_values<-eachCell

  print(expected_values)

  # calculate the Chi Squared test
    chi_squared <- vector("integer", 10)
    for (i in 1:nrow(inputData)){
        for (j in 1:ncol(inputData)){
            chi_squared[i] <- (inputData[i,j] - expected_values[i,j]) ** 2 / expected_values[i,j]
        }
            
    }

        
    print(round(chi_squared,5))

    for (r in 1:nrow(inputData))   
    for (c in 1:ncol(inputData))  
         print(paste("Row", r, "and column",c, "have values of", inputData[r,c]))  
}
```

```python
%%R

chisq(matrix(c(2,3,2,3), nrow = 2, ncol = 2, byrow = T))

```
